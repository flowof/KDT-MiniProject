```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from tqdm.notebook import tqdm
```

```python
url = 'https://comic.naver.com/webtoon/finish'
base_url = 'https://comic.naver.com'
res = requests.get(url)
html_str = res.text
soup = BeautifulSoup(html_str,'html.parser')
a_list = soup.select('div.thumb >a')



webtoon_list = []
for a in a_list:
    title = a.get('title')
    url =base_url + a.get('href')
    thumbnail = a.select_one('img').get('src') 
    get_id = a.select_one('img').get('src').split('/')[5]
    finish = '완결웹툰' 
    
    webtoon_list.append([finish,title,url,thumbnail,get_id])



web_df1 = pd.DataFrame(webtoon_list,columns=['요일','제목','URL','Thumbnail','id'])


web_df1
# web_df1.to_csv('tsttt.csv',encoding='utf-8-sig')

    
```

```python

web_list2 = []

url_list = web_df1['URL'].values.tolist()

for url in tqdm(url_list):

#     html_str = html_cache[url]
    res = requests.get(url)
    html_str = res.text
    soup = BeautifulSoup(html_str,'html.parser')

    author = soup.select('h2 > span.wrt_nm')[0].text.strip()  #작가
    genre = soup.select('span.genre')[0].text #장르
    age_el = soup.select('span.age')
    
    if len(age_el) >=1 :
        age = age_el[0].text
        
    else:
        age = '전체이용가'

    summary= soup.select('.detail>p')[0].text
    id = soup.select_one('img').get('src').split('/')[5]

    
    
    
    web_list2.append([id,author,summary,genre,age])
    
    


web_list2

    
# summary= soup.select('.detail>p')[0].text
# genre = soup.select('span.genre')[0].text
# age = soup.select('span.age')[0].text
```

```python
web_df2 = pd.DataFrame(web_list2,columns=['id','작가','스토리','장르','연령대'])
```

```python
result = pd.merge(web_df1,web_df2, how='inner', on='id')
```

```python
result

id_index = -1
    
for index, name in enumerate(result.columns):
    if name == 'id':
        id_index = index
        break
                
columns = list(result.columns)

columns.pop(id_index)

result = result[['id'] + columns]
```
