- 리스트 딕셔너리 => 데이타프레임 

````python
# 관련 모듈 임포트 
import json 
import pandas as pd
import requests
import urllib.request

# 본인 아이디와 시크릿 키 설정 
client_id = "4qQ1cyzOPJtTPNlD9AL0"
client_secret = "dbaiqb8TMV"

# 검색어 지정후 인코딩 
encText = urllib.parse.quote("부산대학교 맛집")

# url 생성 
# 요청변수 파라미터 지정 
# url 에 ?요청변수1=값1&요청변수2=값2
# 반환되는 블로그 갯수 값 지정 
# ?display=50
# url = "https://openapi.naver.com/v1/search/blog?query=" + encText # json 결과
url = "https://openapi.naver.com/v1/search/blog?display=50&query=" + encText # json 결과
# url = "https://openapi.naver.com/v1/search/blog.xml?query=" + encText # xml 결과
print(url)


# 페이지 요청 
request = urllib.request.Request(url)
request.add_header("X-Naver-Client-Id",client_id)
request.add_header("X-Naver-Client-Secret",client_secret)
response = urllib.request.urlopen(request)
rescode = response.getcode()

# 200 코드이면 response_body에 저장 
if(rescode==200):
    response_body = response.read()
    print(response_body.decode('utf-8'))
else:
    print("Error Code:" + rescode)
````

````python
# json => 딕셔너리
blog = json.loads(response_body.decode('utf-8'))
type(blog), len(blog)
````

````python
# 딕셔너리 => 리스트 딕셔너리 
blog_list = blog['items']
blog_list[0]
````

````python
# 리스트 딕셔너리 => 데이타프레임 
df_blog = pd.DataFrame(columns=['title', 'link', 'description', 
                                'bloggername','bloggerlink', 'postdate' ])
df_blog
````

````python
# 행이름을 생성 후 for 문으로 데이타 값 전달 
idx = 1
for dict in blog_list:
    title = dict['title'].replace('<b>','').replace('</b>','')
    link = dict['link']
    description = dict['description'].replace('<b>','').replace('</b>','')
    bloggername = dict['bloggername']
    bloggerlink = dict['bloggerlink']
    postdate = dict['postdate']
    # 데이타프레임 삽입 
    # 데이타프레임명.loc[행이름] = [값1, 값2....]
    df_blog.loc[idx] = [title, link, description,bloggername,
                        bloggerlink, postdate]
    idx += 1
````

# 퀴즈  : 네이버 책 검색 API 


https://developers.naver.com/docs/serviceapi/search/book/book.md#%EC%B1%85

- 함수로 정의 
- 함수 실행시 검색어(책제목) => ?
- 검색어에 대한 검색 결과가 데이타 프레임으로 출력
- 검색 결과의 갯수는 임의 지정 

```
# 함수 호출 
book_api()

검색어(책제목) => ?

```

````python
def book_api() :
    # 질의키워드 입력받기
    query = urllib.parse.quote(input('검색어 입력 => '))

    # 변수 지정
    display = 100 # 검색건수
    start = 1 # 검색시작위치
    end = 200 # 검색끝위치
    sort = 'sim' # 정렬옵션

    # 데이타프레임 생성
    idx = 0 # 인덱스 변수
    book_df = pd.DataFrame(columns=('Title', 'Link', 'Image', 'Author', 'Price', 'Discount',
                                    'Publisher', 'ISBN', 'Description', 'Pubdate'))

    for start_index in range(start, end, display):
        # 쿼리 URL 생성 => 주소 인코딩
        url = "https://openapi.naver.com/v1/search/book?query=" + query \
              + "&display=" + str(display) \
              + "&start=" + str(start_index) \
              + "&sort=" + sort
        # print('url = ', url)

        # 네이버 API 요청 => 응답
        request = urllib.request.Request(url)
        request.add_header("X-Naver-Client-Id", client_id)
        request.add_header("X-Naver-Client-Secret", client_secret)
        response = urllib.request.urlopen(request)
        rescode = response.getcode()
        # print(rescode)

        # 정상적인 응답 코드
        if (rescode == 200):
            response_body = response.read()

            #  json 객체를 딕셔너리 구조로 변경
            response_dict = json.loads(response_body.decode('utf-8'))
            items = response_dict['items']

            # print(len(items))
            # print(items[0])

            for item_index in range(0, len(items)):
                # 전처리
                title = items[item_index]['title'].replace('<b>', '').replace('</b>', '')
                link = items[item_index]['link']
                description = items[item_index]['description'].replace('<b>','').replace('</b>','')
                image = items[item_index]['image']
                author = items[item_index]['author']
                price = items[item_index]['price']
                discount = items[item_index]['discount']
                publisher = items[item_index]['publisher']
                isbn = items[item_index]['isbn']
                pubdate = items[item_index]['pubdate']

                # 데이타프레임으로 바로 저장
                book_df.loc[idx] = [title, link, image, author, price, discount,
                                    publisher, isbn, description, pubdate]
                idx += 1

        else: # 비정상적인 응답코드
            print("Error Code:" + rescode)

    # 출력
#     print(book_df)
    return book_df

    # csv 파일로 저장
#     book_df.to_csv('output/book_df.csv', index=False)
````

````python
result = book_api()
````
