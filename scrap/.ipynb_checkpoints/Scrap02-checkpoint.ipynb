# 뷰티풀숩을 이용한 작업 과정

- requests/urllib를 이용해서 url 요청 => response 객체 => html화 (텍스트나 바이트)
- html화 된 자료 => 뷰티풀숩 객체화
- 뷰티풀숩의 다양한 함수를 이용해서 자료변환(리스트나 딕셔너리 등)
- csv 파일 등으로 저장 (판다스를 이용)

```python
# 샘플 스크래핑 url 지정
url = 'http://pythonscraping.com/pages/warandpeace.html'
```

```python
#requests를 이용해서 url 요청 => response 객체 => html화 (텍스트)

res = requests.get(url)
print(res) # 샘플 사이즈?가 200이면 잘 맞다?

html_str = res.text
#html_str
```

#### html화 된 자료 => 뷰티풀숩 객체화

- 숩객체명 = BeautifulSoup(html변수 , 해석기)
- 해석기 html.parser/html5/xml/lxml


```python
# 레퍼런스? 참고자료? 물음표 붙이면 그렇대
soup?
```

```python
soup = BeautifulSoup(html_str, 'html.parser')
type(soup)
```

```python
#soup
```

### nextsibling 를 이용한 요소 탐색

- 다음 형제요소 탐색

```python
#dir(soup)
```

```python
print(soup.h1)
print(soup.h1.next_sibling) # 빈 행
print(soup.h1.next_sibling.next_sibling)
print(soup.h1.next_sibling.next_sibling.next_sibling)# 빈 행
```

### find()와 find_all()

- find(옵션) : 1개로 반환
- find_all(옵션) : 리스트로 반환    

```python
# find(태그)
# 태그가 여러개일 경우에는 첫번째 태그만 반환 
print(soup.find('h1'))
print(soup.find('p'))
print(soup.find('span'))
```

```python
# find(id=아이디명)
soup.find(id='text')
```

```python
# find(id=아이디명)
soup.find(id='text').text
```

```python
# find(클래스명)
soup.find(class_= 'green')
```

```python
soup.find(class_= 'red')
```

```python
# find_all(태그) => 리스트 형태
span_list = soup.find_all('span')
len(span_list), type(span_list)
```

```python
#span_list
```

```python
for item in span_list:
    print(item.text)
    print('='*20)
```

```python
# find_all(class_=클래스명)
red_list = soup.find_all(class_='red')
len(red_list)
```

```python
#red_list
```

```python
#red_list[0]
```

```python
# find_all(id=아이디명)
soup.find_all(id='text')
```

```python
# 속성으로 찾기
# find_all(attr={속성:값})
soup.find_all(attrs={'class', 'red'}) 
#soup.find(class_='red') #랑 똑같음
```

### 특정 태그의 속성값 출력
- get(속성명)

```python
green = soup.find(class_='green')
green
```

```python
green.get('class')
```

## select() 와 select_one()

- select(선택자) => 리스트
- select_one(선택자) => 1개의 요소

```python
# red 클래스가 적용된 요소
len(soup.select('.red'))
# len(soup.find_all(class_='red'))
```

```python
#soup.find_all(id='text')
soup.select('#text')
```

```python
# text 아이디값을 갖는 요소 안의 red 클래스를 갖는 span 요소의 리스트
soup.select('#text > span.red')
```
