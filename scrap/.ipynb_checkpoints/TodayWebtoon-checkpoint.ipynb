```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://comic.naver.com/webtoon/weekday.nhn'
res = requests.get(url)
html_str = res.text
res
```

```python
soup = BeautifulSoup(html_str,'html.parser')
```

```python
a_list = soup.select('a.title')
len(a_list)
```

```python
# 한개 요소의 필터링확인
base_url = 'https://comic.naver.com'

title = a_list[0].text
url = base_url + a_list[0].get('href')
weekday = a_list[0].get('href')[-3:]

(title, url, weekday)
```

```python
base_url = 'https://comic.naver.com'
webtoons_list = []
for a in a_list:
    title = a.text
    url = base_url + a.get('href')
    weekday = a.get('href')[-3:]
    webtoons_list.append([weekday,title,url])
```

```python
webtoons_list
```

```python
df_webtoons = pd.DataFrame(webtoons_list,columns=['요일','제목','URL'])
```

```python
df_webtoons
```

```python

# 데이터 프레임의 특정 컬럼의 값 빈도수 확인
df_webtoons['요일'].value_counts()
```

```python
# 데이터프레임의 특정 컬럼의 값 교체
# mon => 월요일, tue = > 화요일 .... sun => 일요일
# 한개의 데이터에 대한 함수 정의
```

```python
def weekChange(x):
    if x == 'mon':
        return '월요일'
    elif x == 'tue':
        return '화요일'
    elif x == 'wed':
        return '수요일'
    elif x == 'thu':
        return '목요일'
    elif x == 'fri':
        return '금요일'
    elif x == 'sat':
        return '토요일'
    elif x == 'sun':
        return '일요일'    
```

```python
weekChange('sun')
```

```python
# 데이터프레임[컬럼명].apply(함수명)
df_webtoons['요일'].apply(weekChange)
```

```python
df_webtoons['요일'] = df_webtoons['요일'].apply(weekChange)
```

```python
df_webtoons[df_webtoons['요일'] == '월요일']
```

```python
df_webtoons.to_csv('output/webbbb.csv',index=False, encoding='cp949')
```

```python
u_list=df_webtoons['URL'].values.tolist()

df_list = []

for i in u_list:

    res = requests.get(i)
    html_str = res.text
    soup = BeautifulSoup(html_str,'html.parser')

    author = soup.select('h2 > span')[0].text.strip()
    summary= soup.select('.detail>p')[0].text
    genre = soup.select('span.genre')[0].text
    age = soup.select('span.age')[0].text
    
    df_list.append([author,summary,genre,age])


dff = pd.DataFrame(df_list,columns=['작가','줄거리','장르','연령제한'])

dff


    

```
