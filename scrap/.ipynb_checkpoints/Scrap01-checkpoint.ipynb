# urllib을 이용한 웹페이지 요청

- 내장 라이브러리 따라서 pip install (x)

```python
# 모듈 임포트
from urllib.request import urlopen
```

```python
url = 'http://google.com'
res = urlopen(url)
res
```

```python
# dir(res)
```

```python
res.code, res.headers
```

#### html 문서 저장

```python
url = 'http://www.aladin.co.kr'
res = urlopen(url)
res.code
```

```python
# res 객체에서 html 소스 저장
html_str = res.read()
type(html_str) # 저장방식이 바이트로 들어가서 아래에 wb로 저장
```

```python
# wb는 byte 저장
url = 'https://www.aladin.co.kr'
res = urlopen(url)
html_str = res.read()
with open('output/alladin.html', 'wb') as f:
    f.write(html_str)
```

```python
ls output
```

```python
html_str2 = str(html_str)
print(type(html_str2))

with open('output/alladin2.html', 'w') as f:
    f.write(html_str2)
```
