# 구글 검색엔진에서 검색어 결과 저장하기

````python
# 1) 관련 모듈 임포트
import pandas as pd
import time #
from selenium import webdriver
import warnings
warnings.filterwarnings(action='ignore')
````

````python
# 2) 셀레니움 드라이버 객체 생성
driver = webdriver.Chrome('chromedriver.exe') # 이 파일을 실행하겠다

# 3) 지연시간 지정 후에 웹브라우저 주소로 이동
time.sleep(2) #2초 지연시간 설정

url = 'http://www.google.com'
driver.get(url)

# 결과적으로 크롬웹브라우저가 자동으로 실행된다
````

````python
# 4) 구글 검색창에 검색어를 입력하고 결과 페이지를 출력하기
# find_element_by_name(name속성값)
# 동작은?
# send_keys(키워드) : 텍스트필드에 값을 입력하는 이벤트
# click() => 선택자 요소 클릭
# submit() => 전송이벤트

'''
from selenium.webdriver.common.by import By

driver.find_element(By.NAME,'q').send_keys('파이썬')
driver.find_element(By.NAME,'q').submit()
'''

driver.find_element_by_name('q').send_keys('파이썬')
driver.find_element_by_name('q').submit()
````

````python
# 5) 지연시간 후에 웹브라우저 닫기

time.sleep(5)
driver.close()
````

### 선택자로 찾기

- find_elements_by_css_selector(선택자) => 리스트로 반환
- find_element_by_css_selector(선택자) => 첫 번째 선택자에 해당하는 것만 반환
- .text => 문자열 반환
- .get_attribute(속성) => 속성에 해당하는 값이 반환

````python
# 크롬 웹브라우저 실행 => 구글 사이트로 이동
# 검색어 입력 => 검색어 결과 페이지 표시

driver = webdriver.Chrome('chromedriver.exe')
time.sleep(2)

url = 'http://www.google.com'
driver.get(url)

driver.find_element_by_name('q').send_keys('파이썬')
driver.find_element_by_name('q').submit()
````

````python
# 첫번째 검색 결과 찾기 
target = 'div.yuRUbf > a > h3'
title = driver.find_element_by_css_selector(target)
print(title)
print(title.text)

target = 'div.yuRUbf > a'
url = driver.find_element_by_css_selector(target)
print(url)
print(url.get_attribute('href'))

target = 'div.IsZvec > div.VwiC3b.yXK7lf.MUxGbd.yDYNvb.lyLwlc.lEBKkf > span'
doc = driver.find_element_by_css_selector(target)
print(doc.text)
````

````python
# 리스트로 반환해서 데이터
````

````python
# 리스트로 반환해서 데이터 프레임으로 저장

# 제목만 리스트로 저장하기
search_list1 = []

target = 'div.yuRUbf > a > h3'
title_list = driver.find_elements_by_css_selector(target)
print(title_list)
print(len(title_list))

for item in title_list:
    search_list1.append(item.text)
    
print(search_list1)
````

````python
# 리스트로 반환해서 데이타 프레임으로 저장 

# 제목만 리스트로 저장하기 
search_list1 = []

target = 'div.yuRUbf > a > h3'
title_list = driver.find_elements_by_css_selector(target)
print(title_list)
print(len(title_list))

for item in title_list:
    search_list1.append(item.text)

print(search_list1)
````

````python
# 하이퍼링크만 리스트로 저장하기 
search_list2 = []

target = 'div.yuRUbf > a'
url_list = driver.find_elements_by_css_selector(target)

for item in url_list:
    search_list2.append(item.get_attribute('href'))

print(search_list2)
print(len(search_list2))
````
