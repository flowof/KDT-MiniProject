```python
# 마지막 페이지까지 댓글 저장 

base_url = 'https://movie.naver.com/movie/point/af/list.naver?st=mcode&sword=13558&target=after&page='
movie_list = []
page = 0 

while True:
    #  (1) url 주소 => html 소스 
    page += 1
    url = base_url+str(page)
    print(url)
    res = requests.get(url)
    html_str = res.text
    
    # (2) soup 객체화 
    soup = BeautifulSoup(html_str, 'html.parser')
    
    # (3) 필요한 요소를 리스트로 저장 
    td_list = soup.select('.title')
    a_list = soup.select('a.author')
    
    
    # (4) 2차원 리스트로 저장 
    for i in range(len(a_list)):
        title = td_list[i].find('a').text
        score = td_list[i].find('em').text
        review = td_list[i].find('br').next_sibling.strip()
        author = a_list[i].text
        date = a_list[i].next_sibling.next_sibling
        movie_list.append([title, score, review, author, date])
    
    # (5) 마지막페이지라면 종료 
    page_next = soup.find_all(class_ = 'pg_next')
    if len(page_next) == 0:
        break

print(f' 전체 댓글 갯수는? {len(movie_list)}')        
```

```python
kid_df = pd.DataFrame(movie_list, 
                        columns = ['영화제목', '평점', '댓글', '작성자', '날짜'])
kid_df
```

### 함수로 코드 개선하기 

```python
def movie_review(url):

    base_url = url
    movie_list = []
    page = 0 

    while True:
        #  (1) url 주소 => html 소스 
        page += 1
        url = base_url+str(page)
        print(url)
        res = requests.get(url)
        html_str = res.text

        # (2) soup 객체화 
        soup = BeautifulSoup(html_str, 'html.parser')

        # (3) 필요한 요소를 리스트로 저장 
        td_list = soup.select('.title')
        a_list = soup.select('a.author')


        # (4) 2차원 리스트로 저장 
        for i in range(len(a_list)):
            title = td_list[i].find('a').text
            score = td_list[i].find('em').text
            review = td_list[i].find('br').next_sibling.strip()
            author = a_list[i].text
            date = a_list[i].next_sibling.next_sibling
            movie_list.append([title, score, review, author, date])

        # (5) 마지막페이지라면 종료 
        page_next = soup.find_all(class_ = 'pg_next')
        if len(page_next) == 0:
            break

    print(f' 전체 댓글 갯수는? {len(movie_list)}')
    field_list = ['영화제목', '평점', '댓글', '작성자', '날짜']
    df = pd.DataFrame(movie_list, columns = field_list)
    return df
```

```python
# 고양이를 부탁해 
url = 'https://movie.naver.com/movie/point/af/list.naver?st=mcode&sword=31607&target=after&page='
cat_df = movie_review(url)
```
