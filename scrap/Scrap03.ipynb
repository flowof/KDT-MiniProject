```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

# 영문 색상명 + 16진수 색상값
- https://www.w3schools.com/colors/colors_names.asp

```python
url = 'https://www.w3schools.com/colors/colors_names.asp'
res = requests.get(url)
print(res)
html_str = res.text
html_str
```

```python
# BeautifulSoup 객체화
soup = BeautifulSoup(html_str, 'html.parser')
#soup
```

```python
# 추출하고자 하는 색상 이름 + 색상코드 1개만 출력
div_list = soup.find_all(class_='innerbox')
len(div_list)
```

```python
div_list[0]
div_list[-1]
```

```python
# 색상명, 16진수색상값, 관련 URL 조회
div_list[0].find(class_='colornamespan').text , div_list[0].find(class_='colorhexspan').text
```

```python
base_url = 'http://www.w3schools.com'
base_url + div_list[0].find('a').get('href')
```

```python
# 모든 색상명, 16진수 색상값, url => 2차원 리스트 구조
color_list = []
base_url = 'http://www.w3schools.com'
for div in div_list:
    color_name = div.find(class_='colornamespan').text
    hex_code = div.find(class_='colorhexspan').text
    url = base_url + div.find('a').get('href')
    color_list.append([color_name, hex_code, url])
```

```python
len(color_list)
```

```python
color_list[-1]
```

```python
# 판다스에서 제공하는 데이터프레임 구조로 변경
# 데이터프레임명 = pd.DataFrame(2차원리스트, columns=필드명리스트)
color_df = pd.DataFrame(color_list, columns=['Name', 'Hex', 'URL'])
```

```python
color_df.head()
```

```python
# csv 파일로 저장
# 데이터프레임명.to_csv(파일경로, index=False)
color_df.to_csv('output/color.csv', index=False)
```
